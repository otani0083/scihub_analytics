```python
scihub2017=pd.read_csv("scihub-2017.statistics.tab",sep='\t',header=None, parse_dates=[0], dtype = {6:'object',7:'object' },chunksize=50000,names=["date","doi","ip","cookie","country","city","longitude","latitude"])
```

```python
def preprocess_japan(x):#countryがjapanのデータを抽出
    x=x[x["country"]=="Japan"]
    return x

df_japan= pd.concat((preprocess_japan(r) for r in scihub2017)) 
```
